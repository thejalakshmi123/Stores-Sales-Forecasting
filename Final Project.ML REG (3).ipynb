# Stores_Sales_Forecasting

```python
import pandas as pd
```

The Stores_Sales_Forecasting dataset contains 2121 rows and 21 columns.This dataset helps businesses in the retail furniture sector analyze past sales to predict future trends, improve inventory management, and make better decisions.

```python
data=pd.read_csv('stores_sales_forecasting.xls',encoding='latin1')
data
```

# LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
a=LabelEncoder()
b=a.fit_transform(data['Order ID'])
c=a.fit_transform(data['Order Date'])
d=a.fit_transform(data['Ship Date'])
e=a.fit_transform(data['Ship Mode'])
f=a.fit_transform(data['Customer ID'])
g=a.fit_transform(data['Customer Name'])
h=a.fit_transform(data['Segment'])
i=a.fit_transform(data['Country'])
j=a.fit_transform(data['City'])
k=a.fit_transform(data['State'])
l=a.fit_transform(data['Region'])
m=a.fit_transform(data['Product ID'])
n=a.fit_transform(data['Category'])
o=a.fit_transform(data['Sub-Category'])
p=a.fit_transform(data['Product Name'])
data['Order ID']=b
data['Order Date']=c
data['Ship Date']=d
data['Ship Mode']=e
data['Customer ID']=f
data['Customer Name']=g
data['Segment']=h
data['Country']=i
data['City']=j
data['State']=k
data['Region']=l
data['Product ID']=m
data['Category']=n
data['Sub-Category']=o
data['Product Name']=p
data
```

# Correlation

```python
data.corr()
```

```python
z=data.drop(['Country','Category'],axis='columns')
z
```

# Train_Test_Split

```python
from sklearn.model_selection import train_test_split
```

```python
x=z.drop('Profit',axis='columns')
x
```

```python
y=z.Profit
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
xtrain,xtest,ytrain,ytest
```

# Linear Regression

```python
from sklearn import linear_model
import warnings
warnings.filterwarnings('ignore')
```

```python
model=linear_model.LinearRegression()
model.fit(xtrain,ytrain)
```

```python
model.predict(xtest)
```

```python
model.score(xtest,ytest)
```

# Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
colors = sns.color_palette('pastel')

z.Region.value_counts().plot(kind="bar", color=colors)

plt.title("Region Count")
plt.xlabel("Region")
plt.ylabel("Count")

plt.show()
```

This plot shows the distribution of data entries across different Regions.One Region has the highest number of entries (over 700), while another has the least (around 350), indicating an imbalance in the dataset. 

```python
sns.countplot(x='Region', hue='Segment', data=z, palette='pastel')
plt.title("Region vs. Segment Count")
plt.xlabel("Region")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

This plot shows how data entries are distributed across different Regions,with each Region further divided by segment categories (0,1, and 2).It highlights which segment is most common in each Region and compares the overall volume of entriess among Regions.

```python
plt.figure(figsize=(10, 6))
plt.hist(z['Sales'], bins=30, color='green', alpha=0.7)
plt.xlabel('Sales Amount')
plt.ylabel('Frequency')
plt.title('Sales Distribution Over Time')
plt.grid()
plt.show()
```

This plot shows the distribution of sales amounts,highlighting how frequently different sales values occur.It reveals that most sales are low in value, while a few are significantly higher.

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Sub-Category', y='Sales', data=z, palette='coolwarm', estimator=sum)
plt.title('Total Sales by Sub-Category')
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

This bar plot visualizes how total sales are distributed among different product sub-categories within a dataset.

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=z, x='Sales', y='Profit', hue='Sub-Category', palette='tab10')
plt.title('Sales vs Profit by Sub-Category')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

This is a scatter plot, and it illustrates the relationship between sales and profit for various product sub-categories.

```python
plt.figure(figsize=(8, 5))
sns.histplot(z['Sales'], bins=15, kde=True, color='teal')
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

This is a distribution plot of sales using histogram combined with a kde(kernel density estimate) curve.It helps to understand how sales values are spread out.
